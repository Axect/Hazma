```python
from hazma import decay
from hazma import parameters
import numpy as np
import matplotlib.pyplot as plt
```

```python
erho = 2 * parameters.rho_mass
es = erho * np.geomspace(1e-6, 1.0, 100)

plt.figure(dpi=150)
plt.plot(es, decay.neutral_rho(es, erho), label=r"$\rho^{0} \to \pi^{\pm} + \pi^{\mp}$")
plt.plot(es, decay.charged_rho(es, erho), label=r"$\rho^{\pm} \to \pi^{\mp} + \pi^{0}$")
plt.yscale("log")
plt.xscale("log")
plt.legend()
plt.xlim(np.min(es), np.max(es))
```
